```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
df = pd.read_csv("Churn.csv", sep=";")
df
```

```python
#separando as variáveis independentes da variável dependente
X = df.drop("Exited", axis=1)
y = df["Exited"]
```

```python
X
```

```python
#deixar os dados numericos na mesma escala
standarscaler = StandardScaler()
numerical = X.select_dtypes(include=['int64','float64']).columns
X[numerical] = standarscaler.fit_transform(X[numerical])
```

```python
X
```

```python
#transformar os valores categoricos em numeros
labelencoder = LabelEncoder()
categorical = X.select_dtypes(include='object').columns
for col in categorical:
  X[col] = labelencoder.fit_transform(X[col])
```

```python
X
```

```python
#dividir dados em treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
#construindo rede neural criando objeto e camadas de neuronios
modelo = Sequential()
modelo.add(Dense(units=64, activation='relu', input_dim=X_treino.shape[1]))#camada de input com nº de colunas dos dados de treino
modelo.add(Dropout(0.4))
modelo.add(Dense(units=32, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=64, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=1, activation='sigmoid'))
```

```python
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
modelo.fit(X_treino, y_treino, epochs=50, batch_size=32)
```

```python
previsoes = modelo.predict(X_teste)
previsoes
```

```python
y_pred = (previsoes > 0.5).astype('int32')
y_pred
```

```python
print('Acurácia: ', accuracy_score(y_teste, y_pred))
print('F1: ', f1_score(y_teste, y_pred))
print('Recall: ', recall_score(y_teste, y_pred))
print('Matriz de Confusão: ', confusion_matrix(y_teste, y_pred))
```
